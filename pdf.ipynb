```python
import camelot
import pandas as pd
```

```python
#abrindo o pdf
tabela = camelot.read_pdf('Grau_De_risco.pdf')
```

```python
#objeto para pandas df
tabela[0].df
```

```python
#plotar todo o texto presente na tabela
camelot.plot(tabela[0], kind='text')
```

```python
#plotar todos os limites da tabela
camelot.plot(tabela[0], kind='contour')
```

```python
#traçar a tabela (para ver se foi detectada corretamente ou não)
camelot.plot(tabela[0], kind='grid')
```

```python
#plotar todos os segmentos de linha da tabela
camelot.plot(tabela[0], kind='line')
```

```python
#plotar todas as interseções da tabela
camelot.plot(tabela[0], kind='joint')
```

```python
#plotar o contorno da tabela
camelot.plot(tabela[0], kind='contour')
```

```python
#flavor = área da tabela que é detectada por padrão
#strip_text = tira os caracteres do texto (nesse caso '\n')
#row_tol = agrupar as linhas (foi testado o número 10 e deu o melhor resultado)

tabela_final = camelot.read_pdf('Grau_De_risco.pdf', flavor='stream', strip_text='\n', row_tol=10)
tabela_saida = tabela_final[0].df
tabela_saida
```

```python
#exportar para .csv
tabela_saida.to_csv('tabela_final.csv', index=False, encoding='latin1', header=False)
```
